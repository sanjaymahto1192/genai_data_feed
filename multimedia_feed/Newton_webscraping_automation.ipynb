---- Function to extract all audio file from provided URL---------------

```python
def extract_audio(soup):
        audio_component = soup.find_all('div',class_='teaser-item teaser-item-1 ms-teaser-item-0 teaser-item-nometa teaser-item-dark teaser-audio')
        model = whisper.load_model('base.en')
        for section in audio_component:
               main_page_link = section.a['href']
               audio_type = section.find('span',class_='tag-title').text
               audio_title = section.h3.text[:15].replace(" ","_") + '.mp3'
               audio_trancribe_file = section.h3.text[:15].replace(" ","_") + '.txt'
               audio_response = requests.get(main_page_link) 
               main_page_soup = BeautifulSoup(audio_response.text, 'html.parser')
               audio_component_source = main_page_soup.find_all('div',class_='full-width-component')
               audio_page_URL = audio_component_source[3].script['src']
               audio_page_response = requests.get(audio_page_URL) 
               audio_soup = BeautifulSoup(audio_page_response.text, 'lxml') 
               audio_source = audio_soup.body.p.style.text.rsplit()[-21][7:-25]
               audio_response = requests.get(audio_source) 

               with open(audio_title, 'wb') as f:
                       f.write(audio_response.content)
                        
               result = model.transcribe(audio_title)
                
               with open(audio_trancribe_file,'w')as file:
                    file.write(result['text'])
```
